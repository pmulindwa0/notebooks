# Web scraping
A python code to web scrap UNABCEC membership list using RequestHTML library

```python
from requests_html import HTMLSession
session = HTMLSession()
import numpy as np
import pandas as pd
```

```python
r = session.get('http://unabcec.co.ug/membership/members_list.php?pagesize=100')
```

```python
gridrow = r.html.find('.bs-gridrow', first = True) #finding the 1st row.
```

```python
#finding all rows and appending them to list 
#and each row data to a dictionary
rows = r.html.find('.bs-gridrow')
database = []
sel_name = '[id$="_organisation_name"]' #CSS selector to select an id that ends with _organisation_name
sel_email = '[id$="_email"]'
sel_contact = '[id$="_contact"]'
sel_postal_address = '[id$="_postal_address"]'
sel_physical_address = '[id$="_physical_address"]'
sel_class = '[id$="_class"]'
for row in rows:
    organ_name = row.find(sel_name, first=True).text
    company_class = row.find(sel_class, first=True).text
    email = row.find(sel_email, first=True).text
    contact = row.find(sel_contact, first=True).text
    postal_address = row.find(sel_postal_address, first=True).text
    phy_address = row.find(sel_physical_address, first=True).text

    data = {
        "name": organ_name,
        "class": company_class,
        "email": email,
        "contact": contact,
        "postal": postal_address,
        "phy_address": phy_address
    }
    database.append(data)
database
```

```python
d = pd.DataFrame(database)
d
```

```python
d.to_csv("UNABCEC.csv", index=False)
```
